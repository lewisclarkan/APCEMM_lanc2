```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'data/uvspec_clear.out'

with open(file_path, 'r') as file:
    data = file.readlines()

for line in data:
    df_clear = pd.DataFrame([x.split() for x in data], columns=['lambda', 'edir', 'edn', 'eup','uavgdir', 'uavgdn', 'uavgup'])

df_clear = df_clear.astype('float')
```

```python
file_path = 'data/uvspec_cloudcover.out'

with open(file_path, 'r') as file:
    data = file.readlines()

for line in data:
    df_cloudcover = pd.DataFrame([x.split() for x in data], columns=['lambda', 'edir', 'edn', 'eup','uavg'])

df_cloudcover = df_cloudcover.astype('float')
```

```python
file_path = 'data/atlas_plus_modtran'

with open(file_path, 'r') as file:
    data = file.readlines()

for line in data:
    df_extra = pd.DataFrame([x.split() for x in data], columns=['lambda', 'e0'])

df_extra = df_extra.astype('float')
```

```python

fig, ax = plt.subplots(1,2, figsize =[7.5,4])

ax[0].plot(df_extra['lambda'], df_extra['e0'])
ax[0].set_xlim([300,340])
ax[0].set_ylim([0,1500])
ax[0].set_xlabel('Wavelength [nm]')
ax[0].set_ylabel('Extraterrestrial irradiation [mW/m$^2$nm]')

ax[1].scatter(df_clear['lambda'], df_clear['edir'], s=10)
ax[1].set_xlabel('Wavelength [nm]')
ax[1].set_ylabel('Direct beam irradiance [mW/m$^2$nm]')

fig.suptitle("Radiative transfer for a clear sky")
fig.tight_layout()
```

```python

fig, ax = plt.subplots(1,2, figsize =[7.5,4])

ax[0].plot(df_extra['lambda'], df_extra['e0'])
ax[0].set_ylim([0,1500])
ax[0].set_xlabel('Wavelength [nm]')
ax[0].set_ylabel('Extraterrestrial irradiation [mW/m$^2$nm]')

ax[1].scatter(df_cloudcover['lambda'], df_cloudcover['edir'], s=10)
ax[1].set_xlabel('Wavelength [nm]')
ax[1].set_ylabel('Direct beam irradiance [mW/m$^2$nm]')

fig.suptitle("Radiative transfer for a clear sky")
fig.tight_layout()
```
